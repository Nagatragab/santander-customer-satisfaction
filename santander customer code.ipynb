```python
import numpy as np

# --- 1. ŸÖÿπÿßŸÑÿ¨ÿ© ÿßŸÑŸÇŸäŸÖÿ© ÿßŸÑÿ¥ÿßÿ∞ÿ© ŸÅŸä 'var3' ---
# ÿ≥ŸÜŸÇŸàŸÖ ÿ®ÿßÿ≥ÿ™ÿ®ÿØÿßŸÑ ÿßŸÑŸÇŸäŸÖÿ© ÿßŸÑÿ∫ÿ±Ÿäÿ®ÿ© -999999 ÿ®ÿßŸÑŸÇŸäŸÖÿ© ÿßŸÑÿ£ŸÉÿ´ÿ± ÿ¥ŸäŸàÿπŸãÿß (the mode)
# ŸàŸáŸä ŸÅŸä Ÿáÿ∞Ÿá ÿßŸÑÿ≠ÿßŸÑÿ© ÿßŸÑŸÇŸäŸÖÿ© 2
train_df['var3'].replace(-999999, 2, inplace=True)
# ŸÑÿß ŸÜÿ≠ÿ™ÿßÿ¨ ŸÑÿ™ÿ∑ÿ®ŸäŸÇŸáÿß ÿπŸÑŸâ test_df ŸÑÿ£ŸÜ .describe() ÿ£ÿ∏Ÿáÿ± ÿ£ŸÜŸáÿß ŸÑÿß ÿ™ÿ≠ÿ™ŸàŸä ÿπŸÑŸâ Ÿáÿ∞Ÿá ÿßŸÑŸÇŸäŸÖÿ©

# --- 2. ÿ•ÿ≤ÿßŸÑÿ© ÿßŸÑÿ≥ŸÖÿßÿ™ ÿßŸÑÿ´ÿßÿ®ÿ™ÿ© ŸàÿßŸÑŸÖŸÉÿ±ÿ±ÿ© ---
# ŸÅÿµŸÑ ŸÖÿ™ÿ∫Ÿäÿ± ÿßŸÑŸáÿØŸÅ (TARGET) ŸàÿßŸÑŸÄ ID
X_train = train_df.drop(columns=['ID', 'TARGET'])
y_train = train_df['TARGET']
X_test = test_df.drop(columns=['ID'])

# ÿ•ÿ≤ÿßŸÑÿ© ÿßŸÑÿ≥ŸÖÿßÿ™ ÿßŸÑÿ´ÿßÿ®ÿ™ÿ©
constant_features = [col for col in X_train.columns if X_train[col].nunique() == 1]
X_train.drop(columns=constant_features, inplace=True)
X_test.drop(columns=constant_features, inplace=True)
print(f"‚úÖ ÿ™ŸÖ ÿ•ÿ≤ÿßŸÑÿ© {len(constant_features)} ÿ≥ŸÖÿ© ÿ´ÿßÿ®ÿ™ÿ©.")

# ÿ•ÿ≤ÿßŸÑÿ© ÿßŸÑÿ≥ŸÖÿßÿ™ ÿßŸÑŸÖŸÉÿ±ÿ±ÿ©
duplicated_mask = X_train.T.duplicated()
duplicated_features = X_train.columns[duplicated_mask].tolist()
X_train.drop(columns=duplicated_features, inplace=True)
X_test.drop(columns=duplicated_features, inplace=True)
print(f"‚úÖ ÿ™ŸÖ ÿ•ÿ≤ÿßŸÑÿ© {len(duplicated_features)} ÿ≥ŸÖÿ© ŸÖŸÉÿ±ÿ±ÿ©.")
print("-" * 30)
print(f"üìä ÿπÿØÿØ ÿßŸÑÿ≥ŸÖÿßÿ™ ÿßŸÑŸÜŸáÿßÿ¶Ÿä ÿ®ÿπÿØ ÿßŸÑÿ™ŸÜÿ∏ŸäŸÅ: {X_train.shape[1]}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

# --- ŸÜŸÅÿ™ÿ±ÿ∂ ÿ£ŸÜ X_train Ÿà y_train ÿ¨ÿßŸáÿ≤ŸàŸÜ ŸÖŸÜ ÿßŸÑÿÆÿ∑Ÿàÿ© ÿßŸÑÿ≥ÿßÿ®ŸÇÿ© ---

# 1. ÿ•ŸÜÿ¥ÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ ÿ£ŸàŸÑŸä ŸÑÿ™ŸÇŸäŸäŸÖ ÿ£ŸáŸÖŸäÿ© ÿßŸÑÿ≥ŸÖÿßÿ™
# n_estimators=100 ŸäÿπŸÜŸä ÿ£ŸÜŸÜÿß ÿ≥ŸÜÿ®ŸÜŸä 100 ÿ¥ÿ¨ÿ±ÿ© ŸÇÿ±ÿßÿ±
# n_jobs=-1 ŸÑÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ŸÉŸÑ ÿ£ŸÜŸàŸäÿ© ÿßŸÑŸÖÿπÿßŸÑÿ¨ Ÿàÿ™ÿ≥ÿ±Ÿäÿπ ÿßŸÑÿπŸÖŸÑŸäÿ©
selector_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
selector_model.fit(X_train, y_train)

# 2. ÿ•ŸÜÿ¥ÿßÿ° ÿ£ÿØÿßÿ© ÿßŸÑÿßÿÆÿ™Ÿäÿßÿ± ÿßŸÑÿ™Ÿä ÿ≥ÿ™ÿ≠ÿ™ŸÅÿ∏ ÿ®ÿßŸÑÿ≥ŸÖÿßÿ™ ÿßŸÑÿ£ŸáŸÖ ŸÅŸÇÿ∑
# threshold='median' ŸäÿπŸÜŸä ÿ£ŸÜŸÜÿß ÿ≥ŸÜÿ≠ÿ™ŸÅÿ∏ ÿ®ÿßŸÑÿ≥ŸÖÿßÿ™ ÿßŸÑÿ™Ÿä ÿ£ŸáŸÖŸäÿ™Ÿáÿß ÿ£ÿπŸÑŸâ ŸÖŸÜ ÿßŸÑŸàÿ≥Ÿäÿ∑
selector = SelectFromModel(selector_model, prefit=True, threshold='median')

# 3. ÿ™ÿ∑ÿ®ŸäŸÇ ÿßŸÑÿßÿÆÿ™Ÿäÿßÿ± ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ® ŸàÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±
X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)

print("‚úÖ ÿ™ŸÖÿ™ ÿπŸÖŸÑŸäÿ© ÿßÿÆÿ™Ÿäÿßÿ± ÿ£ŸÅÿ∂ŸÑ ÿßŸÑÿ≥ŸÖÿßÿ™.")
print("-" * 30)
print(f"ÿπÿØÿØ ÿßŸÑÿ≥ŸÖÿßÿ™ ÿßŸÑÿ£ÿµŸÑŸä: {X_train.shape[1]}")
print(f"üìä ÿπÿØÿØ ÿßŸÑÿ≥ŸÖÿßÿ™ ÿ®ÿπÿØ ÿßŸÑÿßÿÆÿ™Ÿäÿßÿ±: {X_train_selected.shape[1]}")
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# --- ŸÜŸÅÿ™ÿ±ÿ∂ ÿ£ŸÜ y_train, X_train_selected, Ÿà X_test_selected ÿ¨ÿßŸáÿ≤ŸàŸÜ ŸÖŸÜ ÿßŸÑÿÆÿ∑Ÿàÿßÿ™ ÿßŸÑÿ≥ÿßÿ®ŸÇÿ© ---

# 1. ÿ™ŸÇÿ≥ŸäŸÖ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ® ÿ•ŸÑŸâ ŸÖÿ¨ŸÖŸàÿπÿ© ÿ™ÿØÿ±Ÿäÿ® ŸàŸÖÿ¨ŸÖŸàÿπÿ© ÿ™ÿ≠ŸÇŸÇ (Validation)
X_train_part, X_val_part, y_train_part, y_val_part = train_test_split(
    X_train_selected, y_train, test_size=0.25, random_state=42, stratify=y_train
)

# 2. ÿ≠ÿ≥ÿßÿ® ÿßŸÑŸàÿ≤ŸÜ ŸÑŸÖÿπÿßŸÑÿ¨ÿ© ÿπÿØŸÖ ÿ™Ÿàÿßÿ≤ŸÜ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
ratio = (y_train.value_counts()[0]) / (y_train.value_counts()[1])

# 3. ÿ™ÿπÿ±ŸäŸÅ ŸÜŸÖŸàÿ∞ÿ¨ XGBoost
xgb_model = XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    scale_pos_weight=ratio,
    n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    random_state=42,
    n_jobs=-1
)

# 4. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
xgb_model.fit(X_train_part, y_train_part)

print("‚úÖ ÿ™ŸÖ ÿßŸÑÿßŸÜÿ™Ÿáÿßÿ° ŸÖŸÜ ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨!")
print("-" * 30)

# 5. ÿ™ŸÇŸäŸäŸÖ ÿ£ÿØÿßÿ° ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿπŸÑŸâ ŸÖÿ¨ŸÖŸàÿπÿ© ÿßŸÑÿ™ÿ≠ŸÇŸÇ
val_predictions = xgb_model.predict_proba(X_val_part)[:, 1]
auc_score = roc_auc_score(y_val_part, val_predictions)

print(f"üöÄ ŸÜÿ™Ÿäÿ¨ÿ© ÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ (AUC Score) ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿ≠ŸÇŸÇ ŸáŸä: {auc_score:.4f}")
```

```python
from sklearn.model_selection import RandomizedSearchCV

# --- ŸÜŸÅÿ™ÿ±ÿ∂ ÿ£ŸÜ y_train, X_train_selected, Ÿà xgb_model ÿ¨ÿßŸáÿ≤ŸàŸÜ ---

# 1. ÿ™ÿπÿ±ŸäŸÅ ÿ¥ÿ®ŸÉÿ© ÿßŸÑÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿ™Ÿä ŸÜÿ±ŸäÿØ ÿßŸÑÿ®ÿ≠ÿ´ ŸÅŸäŸáÿß
# Ÿáÿ∞Ÿá ŸáŸä "ÿßŸÑÿÆŸäÿßÿ±ÿßÿ™" ÿßŸÑÿ™Ÿä ÿ≥Ÿäÿ¨ÿ±ÿ®Ÿáÿß ÿßŸÑÿ®ÿßÿ≠ÿ´ ÿßŸÑÿ∞ŸÉŸä
param_grid = {
    'n_estimators': [200, 300, 400],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 4, 5, 6],
    'subsample': [0.7, 0.8, 0.9, 1.0] # ŸÜÿ≥ÿ®ÿ© ÿßŸÑÿπŸäŸÜÿßÿ™ ÿßŸÑŸÖÿ≥ÿ™ÿÆÿØŸÖÿ© ŸÅŸä ŸÉŸÑ ÿ¥ÿ¨ÿ±ÿ©
}

# 2. ÿ•ÿπÿØÿßÿØ ÿßŸÑÿ®ÿßÿ≠ÿ´ ÿßŸÑÿπÿ¥Ÿàÿßÿ¶Ÿä (RandomizedSearchCV)
# cv=3: ÿ≥Ÿäÿ≥ÿ™ÿÆÿØŸÖ ÿßŸÑÿ™ÿ≠ŸÇŸÇ ÿßŸÑŸÖÿ™ŸÇÿßÿ∑ÿπ (Cross-Validation) ŸÑÿ™ŸÇŸäŸäŸÖ ŸÉŸÑ ÿ™ŸàŸÑŸäŸÅÿ© ÿ®ÿ¥ŸÉŸÑ ÿ£ÿØŸÇ
# n_iter=20: ÿ≥ŸäÿÆÿ™ÿßÿ± 20 ÿ™ŸàŸÑŸäŸÅÿ© ÿπÿ¥Ÿàÿßÿ¶Ÿäÿ© ŸÖŸÜ ÿßŸÑÿÆŸäÿßÿ±ÿßÿ™ ÿßŸÑŸÖÿ™ÿßÿ≠ÿ©
# scoring='roc_auc': ÿßŸÑŸÖŸÇŸäÿßÿ≥ ÿßŸÑÿ∞Ÿä ÿ≥Ÿäÿ≥ÿ™ÿÆÿØŸÖŸá ŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ£ŸÅÿ∂ŸÑ
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_grid,
    n_iter=20,
    scoring='roc_auc',
    cv=3,
    verbose=1, # ŸÑÿπÿ±ÿ∂ ŸÖÿπŸÑŸàŸÖÿßÿ™ ÿ£ÿ´ŸÜÿßÿ° ÿπŸÖŸÑŸäÿ© ÿßŸÑÿ®ÿ≠ÿ´
    random_state=42,
    n_jobs=-1
)

print("üîç ÿ®ÿØÿ° ÿπŸÖŸÑŸäÿ© ÿßŸÑÿ®ÿ≠ÿ´ ÿπŸÜ ÿ£ŸÅÿ∂ŸÑ ÿ•ÿπÿØÿßÿØÿßÿ™ ŸÑŸÑŸÜŸÖŸàÿ∞ÿ¨...")

# 3. ÿ®ÿØÿ° ÿπŸÖŸÑŸäÿ© ÿßŸÑÿ®ÿ≠ÿ´
random_search.fit(X_train_selected, y_train)

print("\n" + "="*50)
print("üéâ ÿßŸÜÿ™Ÿáÿ™ ÿπŸÖŸÑŸäÿ© ÿßŸÑÿ®ÿ≠ÿ´!")

# 4. ÿπÿ±ÿ∂ ÿ£ŸÅÿ∂ŸÑ ŸÜÿ™Ÿäÿ¨ÿ© Ÿàÿ£ŸÅÿ∂ŸÑ ÿ•ÿπÿØÿßÿØÿßÿ™ ÿ™ŸÖ ÿ•Ÿäÿ¨ÿßÿØŸáÿß
print(f"üèÜ ÿ£ŸÅÿ∂ŸÑ ŸÜÿ™Ÿäÿ¨ÿ© (AUC Score) ÿ™ŸÖ ÿßŸÑÿ™ŸàÿµŸÑ ÿ•ŸÑŸäŸáÿß: {random_search.best_score_:.4f}")
print(f"‚öôÔ∏è ÿ£ŸÅÿ∂ŸÑ ÿ•ÿπÿØÿßÿØÿßÿ™ ŸÑŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸáŸä:")
print(random_search.best_params_)
```

```python
# --- ŸÜŸÅÿ™ÿ±ÿ∂ ÿ£ŸÜ ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ŸÜÿß ÿ¨ÿßŸáÿ≤ÿ© ŸÖŸÜ ÿßŸÑÿÆÿ∑Ÿàÿßÿ™ ÿßŸÑÿ≥ÿßÿ®ŸÇÿ© ---
# X_train_selected, y_train, X_test_selected, test_df

# 1. ÿ™ÿπÿ±ŸäŸÅ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÜŸáÿßÿ¶Ÿä ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿ£ŸÅÿ∂ŸÑ ÿßŸÑÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿ™Ÿä Ÿàÿ¨ÿØŸÜÿßŸáÿß
#    ŸÑÿß ÿ™ŸÜÿ≥Ÿé ÿ•ÿ∂ÿßŸÅÿ© ÿßŸÑÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿ£ÿÆÿ±Ÿâ ÿßŸÑŸÖŸáŸÖÿ© ŸÖÿ´ŸÑ objective Ÿà scale_pos_weight
best_params = {
    'subsample': 0.9,
    'n_estimators': 300,
    'max_depth': 5,
    'learning_rate': 0.05,
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'scale_pos_weight': ratio, # ŸÜÿ≥ÿ™ÿÆÿØŸÖ ŸÜŸÅÿ≥ ÿßŸÑŸÜÿ≥ÿ®ÿ© ÿßŸÑÿ™Ÿä ÿ≠ÿ≥ÿ®ŸÜÿßŸáÿß ÿ≥ÿßÿ®ŸÇŸãÿß
    'random_state': 42,
    'n_jobs': -1
}

final_model = XGBClassifier(**best_params)

# 2. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÜŸáÿßÿ¶Ÿä ÿπŸÑŸâ "ŸÉŸÑ" ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
print("‚è≥ ÿ¨ÿßÿ±Ÿä ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÜŸáÿßÿ¶Ÿä ÿπŸÑŸâ ŸÉŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™...")
final_model.fit(X_train_selected, y_train)
print("‚úÖ ÿßŸÉÿ™ŸÖŸÑ ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÜŸáÿßÿ¶Ÿä!")

# 3. ÿπŸÖŸÑ ÿßŸÑÿ™ŸÜÿ®ÿ§ÿßÿ™ ÿßŸÑŸÜŸáÿßÿ¶Ÿäÿ© ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±
test_predictions = final_model.predict_proba(X_test_selected)[:, 1]

# 4. ÿ•ŸÜÿ¥ÿßÿ° ŸÖŸÑŸÅ ÿßŸÑÿ™ÿ≥ŸÑŸäŸÖ ÿßŸÑŸÜŸáÿßÿ¶Ÿä
submission_df = pd.DataFrame({'ID': test_df['ID'], 'TARGET': test_predictions})
submission_df.to_csv('optimized_submission.csv', index=False)

print("\n" + "="*50)
print("üöÄ ÿ™ŸÖ ÿ•ŸÜÿ¥ÿßÿ° ŸÖŸÑŸÅ ÿßŸÑÿ™ÿ≥ŸÑŸäŸÖ ÿßŸÑŸÜŸáÿßÿ¶Ÿä 'optimized_submission.csv' ÿ®ŸÜÿ¨ÿßÿ≠!")
print("Ÿáÿ∞ÿß ÿßŸÑŸÖŸÑŸÅ ÿ¨ÿßŸáÿ≤ ŸÑŸÑÿ±ŸÅÿπ ÿπŸÑŸâ Kaggle.")
print("ŸÜÿ∏ÿ±ÿ© ÿ≥ÿ±Ÿäÿπÿ© ÿπŸÑŸâ ÿßŸÑŸÖŸÑŸÅ:")
print(submission_df.head())
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

# --- ŸÜŸÅÿ™ÿ±ÿ∂ ÿ£ŸÜ X_train_selected Ÿà y_train ÿ¨ÿßŸáÿ≤ŸàŸÜ ŸÖŸÜ ÿßŸÑÿÆÿ∑Ÿàÿßÿ™ ÿßŸÑÿ≥ÿßÿ®ŸÇÿ© ---

# 1. ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ®ŸÜŸÅÿ≥ ÿßŸÑÿ∑ÿ±ŸäŸÇÿ© ŸÑŸÑŸÖŸÇÿßÿ±ŸÜÿ© ÿßŸÑÿπÿßÿØŸÑÿ©
X_train_part, X_val_part, y_train_part, y_val_part = train_test_split(
    X_train_selected, y_train, test_size=0.25, random_state=42, stratify=y_train
)

# 2. ÿ™ÿπÿ±ŸäŸÅ ŸÜŸÖŸàÿ∞ÿ¨ LightGBM
#    is_unbalance=True ŸáŸà ÿ•ÿπÿØÿßÿØ ÿ±ÿßÿ¶ÿπ ŸÅŸä LightGBM ŸÑŸÖÿπÿßŸÑÿ¨ÿ© ÿπÿØŸÖ ÿ™Ÿàÿßÿ≤ŸÜ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ™ŸÑŸÇÿßÿ¶ŸäŸãÿß
lgbm_model = lgb.LGBMClassifier(
    objective='binary',
    metric='auc',
    is_unbalance=True, 
    n_estimators=1000,         # ŸäŸÖŸÉŸÜŸÜÿß ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿπÿØÿØ ÿ£ŸÉÿ®ÿ± ŸÖŸÜ ÿßŸÑÿ£ÿ¥ÿ¨ÿßÿ± ŸÑÿ£ŸÜŸá ÿ≥ÿ±Ÿäÿπ
    learning_rate=0.05,
    num_leaves=20,             # ŸÑŸÑÿ™ÿ≠ŸÉŸÖ ŸÅŸä ÿ™ÿπŸÇŸäÿØ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
    random_state=42,
    n_jobs=-1,
    colsample_bytree=0.8       # ŸÑÿßÿ≥ÿ™ÿÆÿØÿßŸÖ 80% ŸÖŸÜ ÿßŸÑÿ≥ŸÖÿßÿ™ ŸÅŸä ŸÉŸÑ ÿ¥ÿ¨ÿ±ÿ©
)

# 3. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸÖÿπ ÿÆÿßÿµŸäÿ© ÿßŸÑÿ™ŸàŸÇŸÅ ÿßŸÑŸÖÿ®ŸÉÿ± (Early Stopping)
#    ÿ≥ŸäŸÇŸàŸÖ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿ®ÿßŸÑÿ™ŸàŸÇŸÅ ÿπŸÜ ÿßŸÑÿ™ÿØÿ±Ÿäÿ® ÿ•ÿ∞ÿß ŸÑŸÖ Ÿäÿ™ÿ≠ÿ≥ŸÜ ÿ£ÿØÿßÿ§Ÿá ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿ≠ŸÇŸÇ ŸÑŸÖÿØÿ© 50 ÿØŸàÿ±ÿ©
print("‚è≥ ÿ¨ÿßÿ±Ÿä ÿ™ÿØÿ±Ÿäÿ® ŸÜŸÖŸàÿ∞ÿ¨ LightGBM...")
lgbm_model.fit(X_train_part, y_train_part,
               eval_set=[(X_val_part, y_val_part)],
               eval_metric='auc',
               callbacks=[lgb.early_stopping(50, verbose=False)])

print("‚úÖ ÿßŸÉÿ™ŸÖŸÑ ÿ™ÿØÿ±Ÿäÿ® ŸÜŸÖŸàÿ∞ÿ¨ LightGBM!")

# 4. ÿ™ŸÇŸäŸäŸÖ ÿ£ÿØÿßÿ° ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
val_predictions_lgbm = lgbm_model.predict_proba(X_val_part)[:, 1]
auc_score_lgbm = roc_auc_score(y_val_part, val_predictions_lgbm)

print("\n" + "="*50)
print(f"üöÄ ŸÜÿ™Ÿäÿ¨ÿ© ÿ™ŸÇŸäŸäŸÖ ŸÜŸÖŸàÿ∞ÿ¨ LightGBM (AUC Score) ŸáŸä: {auc_score_lgbm:.4f}")
```
