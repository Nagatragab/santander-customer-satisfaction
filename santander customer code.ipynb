```python
import numpy as np

# --- 1. معالجة القيمة الشاذة في 'var3' ---
# سنقوم باستبدال القيمة الغريبة -999999 بالقيمة الأكثر شيوعًا (the mode)
# وهي في هذه الحالة القيمة 2
train_df['var3'].replace(-999999, 2, inplace=True)
# لا نحتاج لتطبيقها على test_df لأن .describe() أظهر أنها لا تحتوي على هذه القيمة

# --- 2. إزالة السمات الثابتة والمكررة ---
# فصل متغير الهدف (TARGET) والـ ID
X_train = train_df.drop(columns=['ID', 'TARGET'])
y_train = train_df['TARGET']
X_test = test_df.drop(columns=['ID'])

# إزالة السمات الثابتة
constant_features = [col for col in X_train.columns if X_train[col].nunique() == 1]
X_train.drop(columns=constant_features, inplace=True)
X_test.drop(columns=constant_features, inplace=True)
print(f"✅ تم إزالة {len(constant_features)} سمة ثابتة.")

# إزالة السمات المكررة
duplicated_mask = X_train.T.duplicated()
duplicated_features = X_train.columns[duplicated_mask].tolist()
X_train.drop(columns=duplicated_features, inplace=True)
X_test.drop(columns=duplicated_features, inplace=True)
print(f"✅ تم إزالة {len(duplicated_features)} سمة مكررة.")
print("-" * 30)
print(f"📊 عدد السمات النهائي بعد التنظيف: {X_train.shape[1]}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

# --- نفترض أن X_train و y_train جاهزون من الخطوة السابقة ---

# 1. إنشاء نموذج أولي لتقييم أهمية السمات
# n_estimators=100 يعني أننا سنبني 100 شجرة قرار
# n_jobs=-1 لاستخدام كل أنوية المعالج وتسريع العملية
selector_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
selector_model.fit(X_train, y_train)

# 2. إنشاء أداة الاختيار التي ستحتفظ بالسمات الأهم فقط
# threshold='median' يعني أننا سنحتفظ بالسمات التي أهميتها أعلى من الوسيط
selector = SelectFromModel(selector_model, prefit=True, threshold='median')

# 3. تطبيق الاختيار على بيانات التدريب والاختبار
X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)

print("✅ تمت عملية اختيار أفضل السمات.")
print("-" * 30)
print(f"عدد السمات الأصلي: {X_train.shape[1]}")
print(f"📊 عدد السمات بعد الاختيار: {X_train_selected.shape[1]}")
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# --- نفترض أن y_train, X_train_selected, و X_test_selected جاهزون من الخطوات السابقة ---

# 1. تقسيم بيانات التدريب إلى مجموعة تدريب ومجموعة تحقق (Validation)
X_train_part, X_val_part, y_train_part, y_val_part = train_test_split(
    X_train_selected, y_train, test_size=0.25, random_state=42, stratify=y_train
)

# 2. حساب الوزن لمعالجة عدم توازن البيانات
ratio = (y_train.value_counts()[0]) / (y_train.value_counts()[1])

# 3. تعريف نموذج XGBoost
xgb_model = XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    scale_pos_weight=ratio,
    n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    random_state=42,
    n_jobs=-1
)

# 4. تدريب النموذج
xgb_model.fit(X_train_part, y_train_part)

print("✅ تم الانتهاء من تدريب النموذج!")
print("-" * 30)

# 5. تقييم أداء النموذج على مجموعة التحقق
val_predictions = xgb_model.predict_proba(X_val_part)[:, 1]
auc_score = roc_auc_score(y_val_part, val_predictions)

print(f"🚀 نتيجة تقييم النموذج (AUC Score) على بيانات التحقق هي: {auc_score:.4f}")
```

```python
from sklearn.model_selection import RandomizedSearchCV

# --- نفترض أن y_train, X_train_selected, و xgb_model جاهزون ---

# 1. تعريف شبكة الإعدادات التي نريد البحث فيها
# هذه هي "الخيارات" التي سيجربها الباحث الذكي
param_grid = {
    'n_estimators': [200, 300, 400],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 4, 5, 6],
    'subsample': [0.7, 0.8, 0.9, 1.0] # نسبة العينات المستخدمة في كل شجرة
}

# 2. إعداد الباحث العشوائي (RandomizedSearchCV)
# cv=3: سيستخدم التحقق المتقاطع (Cross-Validation) لتقييم كل توليفة بشكل أدق
# n_iter=20: سيختار 20 توليفة عشوائية من الخيارات المتاحة
# scoring='roc_auc': المقياس الذي سيستخدمه لتقييم الأفضل
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_grid,
    n_iter=20,
    scoring='roc_auc',
    cv=3,
    verbose=1, # لعرض معلومات أثناء عملية البحث
    random_state=42,
    n_jobs=-1
)

print("🔍 بدء عملية البحث عن أفضل إعدادات للنموذج...")

# 3. بدء عملية البحث
random_search.fit(X_train_selected, y_train)

print("\n" + "="*50)
print("🎉 انتهت عملية البحث!")

# 4. عرض أفضل نتيجة وأفضل إعدادات تم إيجادها
print(f"🏆 أفضل نتيجة (AUC Score) تم التوصل إليها: {random_search.best_score_:.4f}")
print(f"⚙️ أفضل إعدادات للنموذج هي:")
print(random_search.best_params_)
```

```python
# --- نفترض أن كل بياناتنا جاهزة من الخطوات السابقة ---
# X_train_selected, y_train, X_test_selected, test_df

# 1. تعريف النموذج النهائي باستخدام أفضل الإعدادات التي وجدناها
#    لا تنسَ إضافة الإعدادات الأخرى المهمة مثل objective و scale_pos_weight
best_params = {
    'subsample': 0.9,
    'n_estimators': 300,
    'max_depth': 5,
    'learning_rate': 0.05,
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'scale_pos_weight': ratio, # نستخدم نفس النسبة التي حسبناها سابقًا
    'random_state': 42,
    'n_jobs': -1
}

final_model = XGBClassifier(**best_params)

# 2. تدريب النموذج النهائي على "كل" بيانات التدريب
print("⏳ جاري تدريب النموذج النهائي على كل البيانات...")
final_model.fit(X_train_selected, y_train)
print("✅ اكتمل تدريب النموذج النهائي!")

# 3. عمل التنبؤات النهائية على بيانات الاختبار
test_predictions = final_model.predict_proba(X_test_selected)[:, 1]

# 4. إنشاء ملف التسليم النهائي
submission_df = pd.DataFrame({'ID': test_df['ID'], 'TARGET': test_predictions})
submission_df.to_csv('optimized_submission.csv', index=False)

print("\n" + "="*50)
print("🚀 تم إنشاء ملف التسليم النهائي 'optimized_submission.csv' بنجاح!")
print("هذا الملف جاهز للرفع على Kaggle.")
print("نظرة سريعة على الملف:")
print(submission_df.head())
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

# --- نفترض أن X_train_selected و y_train جاهزون من الخطوات السابقة ---

# 1. تقسيم البيانات بنفس الطريقة للمقارنة العادلة
X_train_part, X_val_part, y_train_part, y_val_part = train_test_split(
    X_train_selected, y_train, test_size=0.25, random_state=42, stratify=y_train
)

# 2. تعريف نموذج LightGBM
#    is_unbalance=True هو إعداد رائع في LightGBM لمعالجة عدم توازن البيانات تلقائيًا
lgbm_model = lgb.LGBMClassifier(
    objective='binary',
    metric='auc',
    is_unbalance=True, 
    n_estimators=1000,         # يمكننا استخدام عدد أكبر من الأشجار لأنه سريع
    learning_rate=0.05,
    num_leaves=20,             # للتحكم في تعقيد النموذج
    random_state=42,
    n_jobs=-1,
    colsample_bytree=0.8       # لاستخدام 80% من السمات في كل شجرة
)

# 3. تدريب النموذج مع خاصية التوقف المبكر (Early Stopping)
#    سيقوم النموذج بالتوقف عن التدريب إذا لم يتحسن أداؤه على بيانات التحقق لمدة 50 دورة
print("⏳ جاري تدريب نموذج LightGBM...")
lgbm_model.fit(X_train_part, y_train_part,
               eval_set=[(X_val_part, y_val_part)],
               eval_metric='auc',
               callbacks=[lgb.early_stopping(50, verbose=False)])

print("✅ اكتمل تدريب نموذج LightGBM!")

# 4. تقييم أداء النموذج
val_predictions_lgbm = lgbm_model.predict_proba(X_val_part)[:, 1]
auc_score_lgbm = roc_auc_score(y_val_part, val_predictions_lgbm)

print("\n" + "="*50)
print(f"🚀 نتيجة تقييم نموذج LightGBM (AUC Score) هي: {auc_score_lgbm:.4f}")
```

```python
# --- نفترض أن كل بياناتنا جاهزة ---
# X_train_selected, y_train, X_test_selected, test_df

# --- 1. تدريب أفضل نموذج XGBoost (باستخدام أفضل الإعدادات التي وجدناها) ---
from xgboost import XGBClassifier

best_xgb_params = {
    'subsample': 0.9, 'n_estimators': 300, 'max_depth': 5,
    'learning_rate': 0.05, 'objective': 'binary:logistic', 'eval_metric': 'auc',
    'scale_pos_weight': ratio, # استخدم النسبة المحسوبة سابقًا
    'random_state': 42, 'n_jobs': -1
}
xgb_final = XGBClassifier(**best_xgb_params)
xgb_final.fit(X_train_selected, y_train)
xgb_predictions = xgb_final.predict_proba(X_test_selected)[:, 1]
print("✅ اكتمل تدريب نموذج XGBoost النهائي.")

# --- 2. تدريب أفضل نموذج LightGBM ---
import lightgbm as lgb

lgbm_final = lgb.LGBMClassifier(
    objective='binary', metric='auc', is_unbalance=True,
    n_estimators=1000, learning_rate=0.05, num_leaves=20,
    random_state=42, n_jobs=-1, colsample_bytree=0.8
)
lgbm_final.fit(X_train_selected, y_train) # تدريبه على كل البيانات
lgbm_predictions = lgbm_final.predict_proba(X_test_selected)[:, 1]
print("✅ اكتمل تدريب نموذج LightGBM النهائي.")

# --- 3. دمج التنبؤات (Ensembling) ---
# سنقوم ببساطة بأخذ المتوسط الحسابي لتنبؤات النموذجين
ensemble_predictions = (xgb_predictions + lgbm_predictions) / 2

# --- 4. إنشاء ملف التسليم النهائي ---
submission_df = pd.DataFrame({'ID': test_df['ID'], 'TARGET': ensemble_predictions})
submission_df.to_csv('ensemble_submission.csv', index=False)

print("\n" + "="*50)
print("🚀 تم إنشاء ملف التسليم النهائي المدمج 'ensemble_submission.csv'!")
print("هذه غالبًا أفضل نتيجة يمكن الوصول إليها.")
print(submission_df.head())
```
